
Librería computación Cuántica: Números complejos
=============

Este proyecto busca el diseño e implementación de una librería en Python que permita realizar operaciones con números complejos. Esta librería, busca ofrecer una herramienta que facilite la manipulación de números complejos tanto en su representación rectangular (tupla con parte real e imaginaria) como en su representación polar (tupla con magnitud y ángulo).


Para Empezar
---------------

### Prerrequisitos

Asegurece de tener instalado:

- Python 3.8+

- git (opcional, para descargar el reposistorio)


### Instalacion


Clonar el repositorio:

    git clone https://github.com/your-username/complex-numbers-library.git
    cd complex-numbers-library


Para correr el codigo con Python:

    from complejos import operaciones
    z1 = (3, 2)   # 3 + 2i
    z2 = (1, -4)  # 1 - 4i

Ejemplo de cómo sacar algunos datos del sistema o usarlos para una pequeña demostración:

    "Suma"
    resultado_suma = suma(z1, z2)
    print("Suma:", resultado_suma)



Ejecutando Las Pruebas
